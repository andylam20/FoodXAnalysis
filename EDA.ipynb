## Amount of items bought per time

```python
items_dict = {item: [] for item in order}
for item in order:
    for time in unique_times:
        count = (data[(data['Time'] == time) & (data['Order'] == item)]['Time'].value_counts().get(time, 0))
        items_dict[item].append(count)
print(items_dict)
```

```python
plt.figure(figsize=(30, 10))
for item in order:
    plt.plot(unique_times, items_dict[item], linestyle='--', marker='s', label=item)
plt.legend()
plt.title('Total Items Ordered per Time')
plt.xlabel('Time (in Military time)')
plt.ylabel('Order Frequencies')
plt.xticks(unique_times)
plt.savefig('Items_Ordered_per_Time.png')
plt.show()
```

# Building a model

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import pickle
```

```python
def transform_input_data(data):
    return pd.get_dummies(data[['Year','Major','University','Time']])
```

```python
def train_model(data):
    X = transform_input_data(data)
    Y = data['Order']

    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
    
    model = HistGradientBoostingClassifier(max_iter=100).fit(X_train, y_train)

    model_filename = 'Xtern_AI.pkl'
    with open(model_filename, 'wb') as model_file:
        pickle.dump(model, model_file)
    
    return model, X_train, X_test, y_train, y_test
```

```python
model, X_train, X_test, y_train, y_test = train_model(data)

y_pred_train = model.predict(X_train)
accuracy = accuracy_score(y_train, y_pred_train)
print(f"Accuracy: {accuracy:.2f}")

with open('Xtern_AI.pkl', 'rb') as model_file:
    new_model = pickle.load(model_file)
    
y_pred = new_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```
